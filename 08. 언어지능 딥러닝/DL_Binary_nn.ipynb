```python
import torch

x_train = torch.FloatTensor([[1,2], [2,3], [3,4], [4,4], [5,3], [6,2], [8,1]])
y_train = torch.FloatTensor([[0], [0], [0], [1], [1], [1], [0]])

nH1 = 10
nH2 = 5
nH3 = 8

W_h1 = torch.randn([2, nH1], requires_grad=True) # 히든 레이어의 W: W_h
b_h1 = torch.randn([nH1], requires_grad=True) # 히든 레이어의 b: b_h
W_h2 = torch.randn([nH1, nH2], requires_grad=True) # 히든 레이어의 W: W_h
b_h2 = torch.randn([nH2], requires_grad=True) # 히든 레이어의 b: b_h
W_h3 = torch.randn([nH2, nH3], requires_grad=True) # 히든 레이어의 W: W_h
b_h3 = torch.randn([nH3], requires_grad=True) # 히든 레이어의 b: b_h

W_o = torch.randn([nH3, 1], requires_grad=True) # 아웃풋 레이어의 W: W_o
b_o = torch.randn([1], requires_grad=True) # 아웃풋 레이어의 b: b_o

optimizer = torch.optim.SGD([W_h1, b_h1, W_h2, b_h2, W_h3, b_h3, W_o, b_o], lr=0.01)
```

## 단계 1: 모델 생성(Setup)

```python
def H(x):
  H1 = torch.sigmoid(torch.matmul(x, W_h1) + b_h1)
  H2 = torch.sigmoid(torch.matmul(H1, W_h2) + b_h2)
  H3 = torch.sigmoid(torch.matmul(H2, W_h3) + b_h3)

  model = torch.sigmoid(torch.matmul(H3, W_o) + b_o) # 아웃풋 레이어 만들기
  return model
```

## 단계 2: 학습(Training)

```python
for iter in range(100000):
  cost = torch.mean((-1)*y_train*torch.log(H(x_train)) - (1-y_train)*torch.log(1-H(x_train)))
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  if iter % 1000 == 0:
    print('[Iteration]', iter, '[Cost]', cost.detach().item())
```

## 단계 3: 추론 수행(Inference)

```python
result = H(x_train)
print(result)
```
