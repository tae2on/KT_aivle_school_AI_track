```python
import torch

x_train = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]])
y_train = torch.FloatTensor([[0], [1], [1], [0]])

nH1 = 3

W_h = torch.randn([2, nH1], requires_grad=True)
b_h = torch.randn([nH1], requires_grad=True)
W_o = torch.randn([nH1, 1], requires_grad=True)
b_o = torch.randn([1], requires_grad=True)
optimizer = torch.optim.SGD([W_h, b_h, W_o, b_o], lr=0.01)
```

## 단계 1: 모델 생성(Setup)

```python
def H(x):
  hidden1 = torch.relu(torch.matmul(x, W_h) + b_h)
  model = torch.sigmoid(torch.matmul(hidden1, W_o) + b_o)
  return model
```

## 단계 2: 학습(Training)

```python
for iter in range(200000):
  cost = torch.mean((-1)*y_train*torch.log(H(x_train)) - (1-y_train)*torch.log(1-H(x_train)))
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  if iter % 10000 == 0:
    print('[Iteration]', iter, '[Cost]', cost.detach().item())
```

## 단계 3: 추론 수행(Inference)

```python
result = H(x_train)
print(result)
```
