```python
import torch

x_train = torch.FloatTensor([[1,2], [2,3], [3,4], [4,4], [5,3], [6,2]])
y_train = torch.FloatTensor([[0], [0], [0], [1], [1], [1]])

W = torch.randn([2, 1], requires_grad=True)
b = torch.randn([1], requires_grad=True)
optimizer = torch.optim.SGD([W, b], lr=0.01)
```

## 단계 1: 모델 생성(Setup)

```python
def H(x):
  model = torch.sigmoid(torch.matmul(x, W) + b)
  return model
```

## 단계 2: 학습(Training)

```python
for iter in range(2000):
  cost = torch.mean((-1)*y_train*torch.log(H(x_train)) - (1-y_train)*torch.log(1-H(x_train)))
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  if iter % 100 == 0:
    print('[Iteration]', iter, '[Cost]', cost.detach().item())
```

## 단계 3: 추론 수행(Inference)

```python
x_test = torch.FloatTensor([[7, 1]])
result = H(x_test)
print('Model Inference with [7,1] ', result.detach().item())

if result.detach().item() < 0.5:
  print('F')
else:
  print('A+')
```
