```python
import torch

x_train = torch.FloatTensor([[1,1], [2,2], [3,3]])
y_train = torch.FloatTensor([[10], [20], [30]])

W = torch.randn([2, 1], requires_grad=True)
b = torch.randn([1], requires_grad=True)
optimizer = torch.optim.SGD([W, b], lr=0.01)
```

## 단계 1: 모델 생성(Setup)

```python
def H(x):
  model = torch.matmul(x, W) + b
  return model
```

## 단계 2: 학습(Training)

```python
for iter in range(2000):
  cost = torch.mean((H(x_train) - y_train)**2)
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()
  if iter % 100 == 0:
    print('[Iteration]', iter, '[Cost]', cost.detach().item())
```

## 단계 3: 추론 수행(Inference)

```python
x_test = torch.FloatTensor([[4, 4]])
result = H(x_test)
print('Model Inference with [4,4] ', result.detach().item())
```
