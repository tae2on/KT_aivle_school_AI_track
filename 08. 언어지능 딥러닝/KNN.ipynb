```python
import numpy as np
group_A = np.zeros((15,2)) # 앞의 값은 샘플 수, 뒤의 값은 샘플의 디멘션 (15행 2열)
group_B = np.zeros((15,2)) # 앞의 값은 샘플 수, 뒤의 값은 샘플의 디멘션 (15행 2열)
group_C = np.zeros((15,2)) # 앞의 값은 샘플 수, 뒤의 값은 샘플의 디멘션 (15행 2열)

# 각 그룹마다 랜덤값을 부여함
# Group A는 100에서 200사이의 실수값
group_A = np.random.uniform(100, 200, (15, 2))
# Group B는 0에서 50사이의 실수 값
group_B = np.random.uniform(0, 50, (15, 2))
# Group C는 x축은 100에서 200사이의 실수값, y축은 0에서 50사이의 실수값
group_C = np.column_stack((np.random.uniform(100, 200, 15), np.random.uniform(0, 50, 15)))

# 각 그룹의 점들과 우리가 가지고 있는 현재 point와의 거리를 구하는 함수가 필요
def distance(point, sample):
  return np.sqrt(np.sum((point-sample)**2))

# knn 함수
def knn(point, k=5):
  distances = []
  for sample in group_A:
    distances.append((distance(point, sample), 'A'))
  for ample in group_B:
    distances.append((distance(point, sample), 'B'))
  for sample in group_C:
    distances.append((distance(point, sample), 'C'))
  distances.sort(key=lambda x: x[0])
  nearest_groups = [group for _, group in distances[:k]]

# 각 그룹의 통계치를 나타내는 리스트
  stat = [nearest_groups.count('A'), nearest_groups.count('B'), nearest_groups.count('C')] # 제일 앞은 k개 중에 몇개가 A에 가장 가까운지에 대한 수치, 두번쨰는 B, 세번째는 C

# 이 stat에서 최대로 큰 값의 위치를 구하면 그게 그룹이 됨
  index = np.argmax(stat)
  if index == 0:
    print('Group A')
  elif index == 1:
    print('Group B')
  else :
    print('Group C')

point = np.array([120, 30])
knn(point)
```
