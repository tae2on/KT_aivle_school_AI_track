```python
data_miso_face = '/content/miso_face'
data_other_face1 = '/content/other_face_1'
data_other_face2 = '/content/other_face_2'
```

```python
import os
import shutil

# ì›ë³¸ í´ë”ë“¤
source_folders = ['miso_face', 'other_face1', 'other_face2']

# í†µí•© ëŒ€ìƒ ê²½ë¡œ
data_face = '/content/face'
data_face_train = os.path.join(data_face, 'train')
data_face_test = os.path.join(data_face, 'test')
data_face_val = os.path.join(data_face, 'valid')

# ê° í´ë” ë‚´ì˜ imagesì™€ labels ê²½ë¡œ
data_folders = {
    'train': {'images': os.path.join(data_face_train, 'images'), 'labels': os.path.join(data_face_train, 'labels')},
    'test': {'images': os.path.join(data_face_test, 'images'), 'labels': os.path.join(data_face_test, 'labels')},
    'valid': {'images': os.path.join(data_face_val, 'images'), 'labels': os.path.join(data_face_val, 'labels')}
}

# í†µí•© í´ë” ìƒì„±
for folder in data_folders.values():
    os.makedirs(folder['images'], exist_ok=True)
    os.makedirs(folder['labels'], exist_ok=True)

# íŒŒì¼ ì˜®ê¸°ê¸°
for source in source_folders:
    for split in ['train', 'test', 'valid']:
        source_images = os.path.join(source, split, 'images')
        source_labels = os.path.join(source, split, 'labels')

        if os.path.exists(source_images):
            for image_file in os.listdir(source_images):
                shutil.copy(os.path.join(source_images, image_file), data_folders[split]['images'])

        if os.path.exists(source_labels):
            for label_file in os.listdir(source_labels):
                shutil.copy(os.path.join(source_labels, label_file), data_folders[split]['labels'])

print("íŒŒì¼ ë³µì‚¬ ì™„ë£Œ")
```
