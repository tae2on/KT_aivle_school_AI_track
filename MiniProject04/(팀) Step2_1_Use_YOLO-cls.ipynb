#### 2) ê° í´ë”ì— ì´ë¯¸ì§€ ë°ì´í„° ì˜®ê¸°ê¸°

```python
import os
import random
import shutil

# ì›ë³¸ ì´ë¯¸ì§€ í´ë” ê²½ë¡œ
source_folder = '/content/my_face'

# ë¶„í•  í›„ ì´ë¯¸ì§€ê°€ ì €ì¥ë  í´ë” ê²½ë¡œ ì„¤ì •
base_dest_folder = '/content/dataset'
train_dest_folder = os.path.join(base_dest_folder, 'train', 'my_face')
val_dest_folder = os.path.join(base_dest_folder, 'val', 'my_face')
test_dest_folder = os.path.join(base_dest_folder, 'test', 'my_face')

# í´ë” ìƒì„± í•¨ìˆ˜
def create_folder(path):
    if not os.path.exists(path):
        os.makedirs(path)

# í•„ìš”í•œ í´ë” ìƒì„±
create_folder(train_dest_folder)
create_folder(val_dest_folder)
create_folder(test_dest_folder)

# ì›ë³¸ í´ë”ì— ìˆëŠ” ì´ë¯¸ì§€ íŒŒì¼ ë¦¬ìŠ¤íŠ¸ ê°€ì ¸ì˜¤ê¸°
image_files = [f for f in os.listdir(source_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]

# ì…”í”Œë§í•˜ì—¬ ì´ë¯¸ì§€ë¥¼ ëœë¤í•˜ê²Œ ì„ìŒ
random.seed(42)  # ì¬í˜„ì„±ì„ ìœ„í•´ seed ì„¤ì •
random.shuffle(image_files)

# ì´ë¯¸ì§€ íŒŒì¼ì„ 8:1:1 ë¹„ìœ¨ë¡œ ë¶„í• 
total_images = len(image_files)
train_split = int(total_images * 0.8)
val_split = int(total_images * 0.1)

train_files = image_files[:train_split]
val_files = image_files[train_split:train_split + val_split]
test_files = image_files[train_split + val_split:]

# ì´ë¯¸ì§€ íŒŒì¼ì„ ê°ê°ì˜ í´ë”ë¡œ ì´ë™
for file in train_files:
    shutil.copy(os.path.join(source_folder, file), os.path.join(train_dest_folder, file))

for file in val_files:
    shutil.copy(os.path.join(source_folder, file), os.path.join(val_dest_folder, file))

for file in test_files:
    shutil.copy(os.path.join(source_folder, file), os.path.join(test_dest_folder, file))

# ë¶„í• ëœ ì´ë¯¸ì§€ ê°œìˆ˜ í™•ì¸
print(f"ì´ ì´ë¯¸ì§€ ê°œìˆ˜: {total_images}")
print(f"Train ì´ë¯¸ì§€ ê°œìˆ˜: {len(train_files)}")
print(f"Validation ì´ë¯¸ì§€ ê°œìˆ˜: {len(val_files)}")
print(f"Test ì´ë¯¸ì§€ ê°œìˆ˜: {len(test_files)}")
```
