#### 1) 본인 얼굴 이미지 데이터셋 불러오기

```python
import os
import zipfile
```

```python
data_myFace = path + '/Datasets/Keras/my_face.zip'
data_myFace
```

```python
## Colab에 생성할 본인 얼굴 폴더 경로
extract_folder = '/content/my_face'

## 위의 경로에 폴더가 없을 때 생성
if not os.path.exists(extract_folder) :
    os.makedirs(extract_folder)

## 위의 경로에 압축을 해제
with zipfile.ZipFile(data_myFace, 'r') as zip_ref :
    file_list = zip_ref.namelist()

    for f in file_list :
        if not f.endswith('/') and f.lower().endswith('.jpg') :
            file_name = os.path.basename(f)

            if not file_name.startswith('._') :
                d_path = os.path.join(extract_folder, file_name)

                with zip_ref.open(f) as source, open(d_path, 'wb') as target :
                    target.write(source.read())
```

#### 2) 각 폴더에 이미지 데이터 옮기기

```python
import os
import random
import shutil

# 원본 이미지 폴더 경로
source_folder = '/content/my_face'

# 분할 후 이미지가 저장될 폴더 경로 설정
base_dest_folder = '/content/dataset'
train_dest_folder = os.path.join(base_dest_folder, 'train', 'my_face')
val_dest_folder = os.path.join(base_dest_folder, 'val', 'my_face')
test_dest_folder = os.path.join(base_dest_folder, 'test', 'my_face')

# 폴더 생성 함수
def create_folder(path):
    if not os.path.exists(path):
        os.makedirs(path)

# 필요한 폴더 생성
create_folder(train_dest_folder)
create_folder(val_dest_folder)
create_folder(test_dest_folder)

# 원본 폴더에 있는 이미지 파일 리스트 가져오기
image_files = [f for f in os.listdir(source_folder) if f.endswith(('.png', '.jpg', '.jpeg'))]

# 셔플링하여 이미지를 랜덤하게 섞음
random.seed(42)  # 재현성을 위해 seed 설정
random.shuffle(image_files)

# 이미지 파일을 8:1:1 비율로 분할
total_images = len(image_files)
train_split = int(total_images * 0.8)
val_split = int(total_images * 0.1)

train_files = image_files[:train_split]
val_files = image_files[train_split:train_split + val_split]
test_files = image_files[train_split + val_split:]

# 이미지 파일을 각각의 폴더로 이동
for file in train_files:
    shutil.copy(os.path.join(source_folder, file), os.path.join(train_dest_folder, file))

for file in val_files:
    shutil.copy(os.path.join(source_folder, file), os.path.join(val_dest_folder, file))

for file in test_files:
    shutil.copy(os.path.join(source_folder, file), os.path.join(test_dest_folder, file))

# 분할된 이미지 개수 확인
print(f"총 이미지 개수: {total_images}")
print(f"Train 이미지 개수: {len(train_files)}")
print(f"Validation 이미지 개수: {len(val_files)}")
print(f"Test 이미지 개수: {len(test_files)}")
```

```python
from PIL import Image

# 이미지 경로 설정
image_path = "/content/drive/MyDrive/test_image3.jpg"
resized_image_path = "/content/drive/MyDrive/test_image3_resized.jpg"

# 이미지 열고 크기 조정
try:
    image = Image.open(image_path)
    resized_image = image.resize((640, 640))  # 이미지 크기를 320x320으로 조정
    resized_image.save(resized_image_path)  # 조정된 이미지 저장
    print(f"Resized image saved at: {resized_image_path}")
except FileNotFoundError:
    print("Image not found. Please check the file path.")
```

```python
# 이미지 리스트에 대해 배치 추론 실행
results = model(["/content/drive/MyDrive/test_image1_resized.jpg", "/content/drive/MyDrive/test_image2_resized.jpg","/content/drive/MyDrive/test_image3_resized.jpg"])  # 결과는 Results 객체들의 리스트로 반환

# 결과 리스트 처리
for result in results:
    boxes = result.boxes  # 경계 상자 출력을 위한 Boxes 객체
    masks = result.masks  # 분할 마스크 출력을 위한 Masks 객체
    keypoints = result.keypoints  # 포즈 출력을 위한 Keypoints 객체
    probs = result.probs  # 분류 출력을 위한 Probs 객체
    obb = result.obb  # OBB 출력을 위한 Oriented Boxes 객체
    result.show()  # 화면에 표시
    result.save(filename="result.jpg")  # 디스크에 저장
```
